```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-3.14/3,3.14/3,0.01)

x2 = np.arange(-5,5,0.01)
plt.figure()
for i in range(np.size(t)):
    print(t)
    x11 = (1- x2* np.cos(2*t[i]))/np.cos(t[i])
    x12 = (-1- x2* np.cos(2*t[i]))/np.cos(t[i])
    plt.plot(x11,x2,'r')
    plt.plot(x12,x2,'b')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
