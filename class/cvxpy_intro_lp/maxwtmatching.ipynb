```python
import numpy as np
import cvxpy as cp 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
W = np.array([[3,2,1],[1,2,3],[2,3,1]])
X = cp.Variable((3,3))
constraints = [X>=0, cp.sum(X,axis=0)==1, cp.sum(X,axis=1)==1]
objective = cp.Maximize( cp.sum(cp.multiply(W,X)))
problem = cp.Problem(objective,constraints)
soln = problem.solve()
print('Max weight = ',soln)
print(X.value)
```

```python
W = np.random.rand(100,100)
X = cp.Variable((100,100))
constraints = [X>=0, cp.sum(X,axis=0)==1, cp.sum(X,axis=1)==1]
objective = cp.Maximize( cp.sum(cp.multiply(W,X)))
problem = cp.Problem(objective,constraints)
soln = problem.solve()
print('Max weight = ',soln)
```
